```python
!pip install requests
import requests
import time

# n8n Webhook URL'inizi buraya girin
N8N_WEBHOOK_URL = "https://senin-render-adresin.onrender.com/webhook/colab-data"

def send_to_n8n(data):
    try:
        response = requests.post(N8N_WEBHOOK_URL, json=data)
        print(f"Sent to n8n: {response.status_code}")
    except Exception as e:
        print(f"Error: {e}")

print("Worker is active. Ready to process heavy tasks.")
```
